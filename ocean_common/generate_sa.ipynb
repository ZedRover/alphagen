```python
import SharedArray as sa
import numpy as np
import glob
```

```python
feature_dirs = glob.glob("/home/public2/share_yw/data/*.npy")
```

```python
feature_names = [i.split("/")[-1].split(".")[0] for i in feature_dirs]
```

```python
for name in feature_names:
    sa.delete(name)
```

```python
for feature_dir in feature_dirs:
    feature_name = "q" + feature_dir.split("/")[-1].split(".")[0]
    feature = np.load(feature_dir)
    print(feature_name, feature.shape)
    try:
        sa.create(feature_name, feature.shape, dtype=np.float32)
    except:
        print("already exist")
    try:
        tmp = sa.attach(feature_name)
        tmp[:] = feature
    except:
        print("already attached")
    print(feature_name, "done")
    print("-" * 20)
```
